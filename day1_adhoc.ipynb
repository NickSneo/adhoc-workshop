```
# loading matplotlib lib
import matplotlib.pyplot as plt
```

```
x=[2,9,3,7]
y=[7,15,5,9]

plt.xlabel("Distance")  #x-axis
plt.ylabel("Time")  #y-axis
plt.plot(x,y,label='railway track')  #draw a line
plt.scatter(x,y,s=200,label='railway station',marker='^')  #plot dots
plt.bar(x,y,label='height')  #plot bar graph
plt.grid(color='red')  #grids
plt.legend()  #show labels
plt.show() #draw graph
```

```
# cricket player data visualization
players=['dhoni','virat','rohit','sir jad']
scores=[120,140,180,89]

# now plot
plt.xlabel('players')
plt.ylabel('scores')
#plt.bar(players,scores,label='2020 score of india')
plt.pie(scores,labels=players,explode=(0.2,0,0,0),shadow='True',autopct='%1.1ff%%')
#plt.grid(color='green')
plt.legend()
plt.show()
```
